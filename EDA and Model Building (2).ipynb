# TRAIN TEST SPLIT

```python
train_data=my_data_downsampled[my_data_downsampled["issue_d"]<"2015-06-01"]
```

```python
# train_data=clean_my_data[clean_my_data["issue_d"]<"2015-06-01"]
```

```python
train_data.shape
```

```python
test_data=my_data_downsampled[my_data_downsampled["issue_d"]>="2015-06-01"]
```

```python
#test_data.shape
```

```python
# test_data=clean_my_data[clean_my_data["issue_d"]>="2015-06-01"]
```

```python
X_train=train_data.drop(["default_ind","issue_d"],axis=1)
```

```python
X_train.shape
```

```python
Y_train=train_data["default_ind"]
```

```python
X_test=test_data.drop(["default_ind","issue_d"],axis=1)
X_test.shape
```

```python
Y_test=test_data["default_ind"]
```

```python
from sklearn.tree import ExtraTreeClassifier
selection=ExtraTreeClassifier()
```

```python

selection.fit(X_train,Y_train)
```

```python
selection.feature_importances_
```

```python
importance_values=(selection.feature_importances_)*100
```

```python
importance_values
```

```python
columns_data_imp=X_train.columns
```

```python
importance_data=pd.DataFrame({"columns_data_imp":columns_data_imp,"importance_values":importance_values})
importance_data
```

```python
sorted_importance= importance_data.sort_values(by='importance_values',ascending=False)
sorted_importance
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic_classifier= LogisticRegression()
logistic_classifier.fit(X_train,Y_train)
```

```python
Y_pred = logistic_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
cfm = confusion_matrix(Y_test, Y_pred)
print(cfm)
```

```python
acc= accuracy_score(Y_test, Y_pred)
```

```python
acc
```

```python
print(classification_report(Y_test,Y_pred))
```

```python

p = sns.heatmap(pd.DataFrame(cfm), annot=True, cmap="magma" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = logistic_classifier.predict_proba(X_test)[:,1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
legend_data=roc_auc_score(Y_test,y_pred_proba)
legend_data=round(legend_data,4)
legend_data
```

```python
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_proba)
print('FPR')
print(fpr)
print('TPR')
print(tpr)
print('Thresholds')
print(thresholds)
```

```python
plt.style.use('ggplot')
plt.plot(fpr,tpr, label='logistic regression',color='orange')
plt.plot([0,1],[0,1],'k--',color='navy')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title(' ROC curve')
plt.legend(["AUC=0.9345"], loc ="lower right") 
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(logistic_classifier, X_train, Y_train, cv=10, scoring = "accuracy")
print("Scores:", scores)
print("Mean:", scores.mean())
```
